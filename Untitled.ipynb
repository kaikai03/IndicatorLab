```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
import scipy.stats as st
import statsmodels.api as sm

from sklearn import linear_model
```

```python

def make_half_life_ind(period):
    return np.exp(np.log(0.5)/period)

def weight_half_life(T,period):
    '''半衰期权重生成：T 当前轮数， period：半衰期长度'''
    return round(np.power(make_half_life_ind(50), 50),8)

def neutralize(factor:pd.Series, data, categorical:list=None, logarithmetics:list=None):
    '''市中性化：'''    
    X = data    
    # 对数化
    if not logarithmetics is None:
        X[logarithmetics] = np.log(X[logarithmetics])
    # 哑变量
    if not categorical is None:
        X = pd.get_dummies(X,categorical)

    model = linear_model.LinearRegression().fit(X, factor)
    factor_fitted = factor - model.predict(X)
    neutralize_factor = factor - factor_fitted

    return neutralize_factor

def get_rank_ic(factor, ret_forward):
    """
       计算因子的信息系数
    输入：
        factor: 因子值, multiIndex[date,code]
        ret_forward: 下一期的股票收益率,multiIndex[date,code]
    返回：
        DataFrame:包含['IC']和['pValue']
    """
    
    assert len(factor.index.get_level_values(1).unique()) > 4, '股票数量必须大于4，否则没啥意义啊'
    
    common_index = factor.index.get_level_values(0).unique().intersection(ret_forward.index.get_level_values(0).unique())
    ic_data = pd.DataFrame(index=common_index, columns=['IC','pValue'])

    # 计算相关系数
    for dt in ic_data.index:
        tmp_factor = factor[dt]
        tmp_ret = ret_forward[dt]
   
        df = pd.DataFrame({'factor':tmp_factor, 'ret_f':tmp_ret},index=tmp_factor.index.intersection(tmp_ret.index).to_list())
#         print(dt)
#         print(df)
        df = df[~pd.isnull(df['factor'])][~pd.isnull(df['ret_f'])]
        if len(df) < 5:
            continue

#         ic, p_value = st.spearmanr(df['factor'], df['ret_f'])   # 计算秩相关系数RankIC
#         ic_data['IC'][dt] = ic
#         ic_data['pValue'][dt] = p_value

#         ic_data['IC'][dt] = df['factor'].corr(df['ret_f'],method='spearman')

        ic = df['factor'].rank().corr(df['ret_f'].rank(),method='pearson')
        
        ic_data['IC'][dt] = ic
           
    return ic_data

def get_ic_desc(ic_data):
    mean = ic_data.mean()
    std = ic_data.std()
    t = mean / (std/np.sqrt(len(ic_data)-1))  
    tt,p = st.ttest_ind(ic_data.dropna().tolist(), [0] * len(ic_data))
    return mean,std,t,tt,p
    
def get_ic_ir(ic_data):
    return ic_data.mean()/ic_data.std()
```
